#Write a program to demonstrate how LU decomposition help in solving linear systems?

```python
import numpy as np
import scipy.linalg
A=np.array([[2,-1,0,0],
             [-1,2,-1,0],
             [0,-1,2,-1],
             [0,0,-1,2]])
b=np.array([3,-2,-2,3])
P,L,U=scipy.linalg.lu(A)
print("Lower Triangular matrix",L)
print("Upper Triangular matrix",U)
print("Permutation matrix",P)
x1=np.linalg.solve(A,b)
print("solution",x1)
error=np.dot(A,x1)-b
print("Error",error)
```

#How do different solution methods compare in computational efficiency?

```python
import numpy as np
import sympy as sp
import time
Num=50
A=np.random.rand(Num,Num)+Num*np.eye(Num)
b=np.random.rand(Num,1)
start_time=time.time()
x1=np.linalg.solve(A,b)
end_time=time.time()
print("LU",x1)
print("Time taken for LU",end_time-start_time)
start_time=time.time()
x2=np.linalg.inv(A)@b
end_time=time.time()
print("Inverses",x2)
print("Time taken for Inverse",end_time-start_time)
start_time=time.time()
C=np.hstack((A,b))
C=sp.Matrix(C)
D,_=C.rref()
R=np.array(D).astype(np.float64)
x3=R[:-1]
end_time=time.time()
print("RRef",x3)
print("Time taken for RREf",end_time-start_time)
```

#Write a program to construct Colley’s matrix and right-hand side


```python
import numpy as np
from scipy.io import loadmat
data_scores=loadmat("/content/Scores.mat")
Scores=data_scores["Scores"]
games=np.abs(Scores)
total=np.sum(games,axis=1)
ColleyMatrix=2*np.eye(10)+np.diag(total)-games
RightSide=1+0.5*np.sum(Scores,axis=1)
print("ColleyMatrix",ColleyMatrix)
print("Right Side",RightSide)
```

#Write a program to solve Colley’s system of linear equations.

```python

RanksColley=np.linalg.solve(ColleyMatrix,RightSide)
print("Ranks:",RanksColley)
```
